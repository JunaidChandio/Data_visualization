# seaborn_heatmap

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
arr_2d=np.linspace(1,5,12).reshape(4,3)
arr_2d
```

```python
sns.heatmap(arr_2d)
```

```python
global_warming_df=pd.read_csv("C:\\Users\\Junaid Ahmed\\Downloads\\Who_is_responsible_for_global_warming.csv")
global_warming_df
```

```python
globalwarming_df=global_warming_df.drop(columns=['Country Code','Indicator Name','Indicator Code']).set_index("Country Name")
globalwarming_df.head(6)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalwarming_df)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalwarming_df,vmin=0,vmax=21)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalwarming_df,cmap="coolwarm")
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalwarming_df,cmap="coolwarm",annot=True)
```

```python
plt.figure(figsize=(16,9))
annot_kws={"fontsize":10,"fontstyle":"italic","color":"k","alpha":0.6,"rotation":"vertical","verticalalignment":"center","backgroundcolor":"w"}
sns.heatmap(globalwarming_df,cmap="coolwarm",annot=True,annot_kws=annot_kws)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalwarming_df,cmap="coolwarm",annot=True,linewidths=4)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalwarming_df,cmap="coolwarm",annot=True,linewidths=4,linecolor="k")
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalwarming_df,cbar=False)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalwarming_df,cbar=False,xticklabels=False,yticklabels=False)
```

```python
plt.figure(figsize=(16,9))
cbar_kws={"orientation":"horizontal","shrink":1,"extend":"min","extendfrac":0.1,"ticks":np.arange(0,22),"drawedges":True}
sns.heatmap(globalwarming_df,cbar_kws=cbar_kws)
```

```python
plt.figure(figsize=(16,9))
country_code=['USA','GBR','IND','CHN','RUS','AUS','FRA','DEU','CAN','BRA','ARG','PAK','NPL','BGD','JPN']
sns.heatmap(globalwarming_df,xticklabels=np.arange(1,15),yticklabels=country_code)
```

```python
plt.figure(figsize=(16,9))

ax=sns.heatmap(globalwarming_df)
ax.set(title="who is responsible for globalwarmin",
      xlabel="COZ emission (metric ton per capita)per year",
      ylabel="country name")
```

# heatmap_correlation

```python
globalwarming_df.corr()
```

```python
plt.figure(figsize=(15,8))
ax=sns.heatmap(globalwarming_df.corr(),annot=True,linewidths=2)
ax.tick_params(size=10,color="w",labelsize=15,labelcolor="k")
plt.title('heatmap of who is responsible for globalwarming')
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
cancer_dataset=load_breast_cancer()
cancer_dataset
```

```python
cancer_df=pd.DataFrame(np.c_[cancer_dataset['data'],cancer_dataset['target']],
columns=np.append(cancer_dataset['feature_names'],['target']))
cancer_df
```

```python
plt.figure(figsize=(25,25))
ax=sns.heatmap(cancer_df.corr(),annot=True,linewidths=3)
ax.tick_params(size=1,color="w",labelsize=12,labelcolor='k')
plt.title("heatmap of breast cancer patient")
plt.show()
```
