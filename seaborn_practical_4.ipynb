# seaborn scatter

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_df=sns.load_dataset("titanic")
```

```python
titanic_df
```

```python
sns.scatterplot(x="age",y="fare",data=titanic_df)
```

```python
sns.scatterplot(x="age",y="fare",data=titanic_df,hue="sex")
```

```python
sns.scatterplot(x="age",y="fare",data=titanic_df,hue="sex",style="who")
```

```python
sns.scatterplot(x="age",y="fare",data=titanic_df,hue="sex",style="who",size="who")
```

```python
sns.scatterplot(x="age",y="fare",data=titanic_df,hue="sex",style="who",size="who",sizes=(100,500))
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x="age",y="fare",data=titanic_df,hue="sex",style="who",size="who",sizes=(100,500))
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x="who",y="fare",data=titanic_df,hue="alive",style="alive",size="who",sizes=(100,500))
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x="who",y="fare",data=titanic_df,hue="alive",style="alive",size="who",sizes=(100,500),palette="inferno_r",alpha=0.7)
```
