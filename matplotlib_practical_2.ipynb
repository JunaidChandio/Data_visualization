# Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_student_age=np.random.randint(18,45,(100))
py_student_age=np.random.randint(15,40,(100))
```

```python
print(ml_student_age)
print(py_student_age)
```

```python
plt.hist(ml_student_age)
plt.title("ML students age histogram")
plt.xlabel("students age category")
plt.ylabel("No. students age")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8)
plt.title("ML students age histogram")
plt.xlabel("students age category")
plt.ylabel("No. students age")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8,histtype='step')
plt.title("ML students age histogram")
plt.xlabel("students age category")
plt.ylabel("No. students age")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8,histtype='bar',orientation='horizontal')
plt.title("ML students age histogram")
plt.xlabel("students age category")
plt.ylabel("No. students age")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8,histtype='bar',color='m',label='Ml students')
plt.title("ML students age histogram")
plt.xlabel("students age category")
plt.ylabel("No. students age")
plt.legend(loc=2)
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.figure(figsize=(14,5))
plt.hist([ml_student_age,py_student_age],bins,rwidth=0.8,histtype='bar',color=['m','y'],label=['Ml students','py students'])
#plt.hist(py_student_age,bins,rwidth=0.8,histtype='bar',color='g',label='py students')
plt.title("ML and Python students age histogram")
plt.xlabel("students age category")
plt.ylabel("No. students age")
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
bins=[15,20,25,30,35,40,45]
plt.figure(figsize=(14,5))
plt.hist([ml_student_age,py_student_age],bins,rwidth=0.8,histtype='bar',color=['m','y'],label=['Ml students','py students'])
plt.title("ML and Python students age histogram")
plt.xlabel("students age category")
plt.ylabel("No. students age")
plt.show()
```
