# bar chart

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes=['python','R','AI','ML','DS']
class1_students=[30,10,20,25,10]
class2_students=[40,5,20,20,10]
class3_students=[35,5,30,15,15]
```

```python
plt.bar(classes,class1_students)
```

```python
plt.barh(classes,class1_students)
```

```python
plt.bar(classes,class1_students,width=0.3)
```

```python
plt.bar(classes,class1_students,width=0.3,align='edge',color='y',edgecolor='m',linewidth=5,alpha=0.9)
```

```python
plt.bar(classes,class1_students,width=0.3,align='edge',color='y',edgecolor='m',linewidth=5,alpha=0.9,linestyle=':')
```

```python
plt.bar(classes,class1_students,width=0.3,align='edge',color='y',edgecolor='m',linewidth=5,alpha=0.9,linestyle=':',label='class1 students',visible=False)
```

```python
plt.bar(classes,class1_students,width=0.3,align='edge',color='y',edgecolor='m',linewidth=5,alpha=0.9,linestyle=':',label='class1 students',visible=True)
```

```python
style.use('ggplot')
plt.figure(figsize=(10,5))
plt.bar(classes,class1_students,width=0.4,align='edge',color='k',edgecolor='m',linewidth=5,alpha=0.9,linestyle=':',
        label='class1 students',visible=True)
```

```python
plt.figure(figsize=(10,5))
plt.barh(classes,class1_students,align='edge',color='k',edgecolor='m',linewidth=5,alpha=0.9,linestyle=':',
        label='class1 students',visible=True)
```

```python
plt.figure(figsize=(10,5))
plt.bar(classes,class1_students,width=0.6,align='edge',color='k',edgecolor='m',linewidth=5,alpha=0.9,linestyle=':',
        label='class1 students',visible=True)
plt.title('bar chart of IAIP class',fontsize=20)
plt.xlabel('classes',fontsize=15)
plt.ylabel("no.of students",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(classes,class1_students,width=0.2,color='b',label='class1 students')
plt.bar(classes,class2_students,width=0.2,color='g',label='class2 students')
plt.title('bar chart of IAIP class',fontsize=20)
plt.xlabel('classes',fontsize=15)
plt.ylabel("no.of students",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10,5))
class_index=np.arange(len(classes))
width=0.2
plt.bar(class_index,class1_students,width,color='b',label='class1 students')
plt.bar(class_index+width,class2_students,width,color='g',label='class2 students')
plt.xticks(class_index+width,classes,rotation=15)
plt.title('bar chart of IAIP class',fontsize=20)
plt.xlabel('classes',fontsize=15)
plt.ylabel("no.of students",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10,5))
class_index=np.arange(len(classes))
width=0.2
plt.bar(class_index,class1_students,width,color='b',label='class1 students')
plt.bar(class_index+width,class2_students,width,color='g',label='class2 students')
plt.bar(class_index+width+width,class3_students,width,color='y',label='class2 students')
plt.xticks(class_index+width,classes,rotation=15)
plt.title('bar chart of IAIP class',fontsize=20)
plt.xlabel('classes',fontsize=15)
plt.ylabel("no.of students",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10,5))
class_index=np.arange(len(classes))
width=0.2
plt.barh(class_index,class1_students,width,color='b',label='class1 students')
plt.barh(class_index+width,class2_students,width,color='g',label='class2 students')
plt.barh(class_index+width+width,class3_students,width,color='y',label='class3 students')
plt.yticks(class_index+width,classes,rotation=15)
plt.title('bar chart of IAIP class',fontsize=20)
plt.ylabel('classes',fontsize=15)
plt.xlabel("no.of students",fontsize=15)
plt.legend()
plt.show()
```
