# Pie chart

```python
import matplotlib.pyplot as plt
```

```python
classes=["python","R","Machine learning","Artificial intelligence","Data Science"]
class1_students=[40,15,35,25,30]
plt.pie([1])
plt.show()
```

```python
#classes=["python","R","Machine learning","Artificial intelligence","Data Science"]
#class1_students=[40,15,35,25,30]
plt.pie(class1_students,labels=classes)
plt.show()
```

```python
explode=[0.03,0,0.1,0,0]
plt.pie(class1_students,labels=classes,explode=explode)
plt.show()
```

```python
colors=['b','c','r','y','g']
explode=[0.03,0,0.1,0,0]
plt.pie(class1_students,labels=classes,explode=explode,colors=colors)
plt.show()
```

```python
colors=['b','c','r','y','g']
explode=[0.03,0,0.1,0,0]
plt.pie(class1_students,labels=classes,explode=explode,colors=colors,autopct="%0.2f%%",shadow=True)
plt.show()
```

```python
colors=['b','c','r','y','g']
explode=[0.03,0,0.1,0,0]
plt.pie(class1_students,labels=classes,explode=explode,colors=colors,autopct="%0.2f%%",shadow=True,radius=1.1,startangle=180)
plt.show()
```

```python
colors=['b','c','r','y','g']
textprops={"fontsize":15}
explode=[0.03,0,0.1,0,0]
plt.pie(class1_students,labels=classes,explode=explode,colors=colors,autopct="%0.2f%%",shadow=True,startangle=90,textprops=textprops)
plt.show()
```

```python
colors=['b','c','r','y','g']
textprops={"fontsize":15}
explode=[0.03,0,0.1,0,0]
plt.pie(class1_students,labels=classes,
        explode=explode,colors=colors,autopct="%0.2f%%",
        shadow=True,startangle=90,textprops=textprops,
       pctdistance=0.7,labeldistance=1.2,counterclock=True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
wedgeprops={'linewidth':6,'width':1,'edgecolor':'k'}
colors=['b','c','r','y','g']
textprops={"fontsize":15}
explode=[0.03,0,0.1,0,0]
plt.pie(class1_students,labels=classes,
        explode=explode,colors=colors,autopct="%0.2f%%",
        shadow=True,startangle=270,textprops=textprops,
       pctdistance=0.6,labeldistance=1.4,counterclock=True,
       wedgeprops=wedgeprops,center=(2,3),frame=True,rotatelabels=True,radius=1.1)
plt.show()
```

```python
plt.figure(figsize=(15,8))
wedgeprops={'linewidth':6,'width':1,'edgecolor':'k'}
colors=['b','c','r','y','g']
textprops={"fontsize":15}
explode=[0.03,0,0.1,0,0]
plt.pie(class1_students,labels=classes,
        explode=explode,colors=colors,autopct="%0.2f%%",
        shadow=True,startangle=270,textprops=textprops,
       pctdistance=0.6,labeldistance=1.4,counterclock=True,
       wedgeprops=wedgeprops,center=(2,3),frame=True,rotatelabels=False,radius=1.1,)
plt.legend(loc=2)
plt.show()
```
