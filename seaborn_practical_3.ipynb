```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips_df=sns.load_dataset("tips")
tips_df
```

```python
sns.barplot(tips_df.total_bill)
```

```python
sns.barplot(x=tips_df.day,y=tips_df.total_bill)
```

```python
sns.barplot(x=tips_df.day,y=tips_df.total_bill,hue=tips_df.sex)
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df)
```

```python
order=['Sun',"Thur","Fri","Sat"]
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,order=order)
```

```python
order=['Sun',"Thur","Fri","Sat"]
hue_order=['Female','Male']
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,order=order,hue_order=hue_order)
```

```python

sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,estimator=np.mean)
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,estimator=np.max)
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,ci=20)
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,estimator=np.max,n_boot=2)
```

```python
sns.barplot(y='day',x='total_bill',hue='sex',data=tips_df,estimator=np.max,n_boot=6)
```

```python
sns.barplot(x="total_bill",y="size",hue="sex",data=tips_df,orient="h")
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,color="g")
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,palette="hot",saturation=.8)
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,errcolor="0.5")
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,errwidth=12)
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,capsize=0.1)
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,dodge=False)
```

```python
sns.set()
kwargs={"alpha":0.9,'linestyle':":","linewidth":5,"edgecolor":"k"}
sns.barplot(x='day',y='total_bill',data=tips_df,**kwargs)
```

```python
ax=sns.barplot(x='day',y='total_bill',data=tips_df,alpha=0.9,linestyle="-.",linewidth=3,edgecolor="g")
ax.set(title="barplot of tips dataframe",
      xlabel="days",
      ylabel="total bill")
```

```python
plt.figure(figsize=(16,9))
sns.barplot(x='day',y='total_bill',data=tips_df,alpha=0.9,linestyle="-.",linewidth=3,edgecolor="k")
plt.title("bar plot of days and total billt",fontsize=20)
plt.xlabel("days",fontsize=15)
plt.ylabel("total bill",fontsize=15)
plt.savefig("bar plot of days and total bill")
plt.show()
```
