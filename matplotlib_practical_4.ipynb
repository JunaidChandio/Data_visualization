```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_google_play_store_apps=pd.read_csv("C:\\Users\\Junaid Ahmed\\Downloads\\googleplaystore.csv",nrows=1000)
df_google_play_store_apps.shape
```

```python
df_google_play_store_apps
```

```python
x=df_google_play_store_apps["Rating"]
y=df_google_play_store_apps["Reviews"]
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.title("google play store app scatter")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,c="r",marker="*",s=9,alpha=(0.2),linewidths=8,edgecolors='m')
plt.title("google play store app scatter")
plt.xlabel("Rating")
plt.ylabel("Reviews and Installs")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,c="r",marker="*",s=9,alpha=(0.2),linewidths=8,edgecolors='m',)
plt.title("google play store app scatter")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,c="r",marker="*",s=90,alpha=(0.8),linewidths=8,edgecolors='g',)
plt.scatter(x,df_google_play_store_apps["Installs"],c="y",marker="o",s=90,alpha=(0.8),linewidths=8,edgecolors='c',)
plt.title("google play store app scatter")
plt.xlabel("Rating")
plt.ylabel("Reviews and Installs")
```
