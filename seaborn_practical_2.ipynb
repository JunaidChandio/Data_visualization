```python
tips_df=sns.load_dataset("tips")
tips_df
```

```python
sns.distplot(tips_df["size"])
```

```python
sns.distplot(tips_df["total_bill"],label="total bill")
plt.title("histogram of total bill")
plt.legend()
plt.show()
```

```python
sns.set()
sns.distplot(tips_df["total_bill"],label="total bill")
plt.title("histogram of total bill")
plt.legend()
plt.show()
```

```python
tips_df.total_bill.sort_values()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
sns.set()
plt.figure(figsize=(15,8))
sns.distplot(tips_df["total_bill"],bins=bins,)
plt.xticks(bins)
plt.title("histogram of total bill")
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.distplot(tips_df["total_bill"],hist_kws={"color":"b","edgecolor":"m","linewidth":5,"linestyle":"--","alpha":0.9},
             kde_kws={"color":"r","linewidth":5,"linestyle":"--","alpha":0.9},
            rug=True)
```

```python
plt.figure(figsize=(15,8))
sns.distplot(tips_df["total_bill"],hist_kws={"color":"b","edgecolor":"m","linewidth":5,"linestyle":"--","alpha":0.9},
             kde_kws={"color":"r","linewidth":5,"linestyle":"--","alpha":0.9},
            rug=True,
            rug_kws={"color":"k","edgecolor":"m","linewidth":5,"linestyle":"--","alpha":0.9})
```

```python
plt.figure(figsize=(15,8))
sns.distplot(tips_df["total_bill"],hist_kws={"color":"b","edgecolor":"m","linewidth":5,"linestyle":"--","alpha":0.9},
             kde=False,
             fit=norm,
             fit_kws={"color":"y","linewidth":5,"linestyle":"--","alpha":0.9},
            rug=True,
            rug_kws={"color":"k","edgecolor":"m","linewidth":5,"linestyle":"--","alpha":0.9})
```

```python
plt.figure(figsize=(15,8))
sns.distplot(tips_df["total_bill"],hist_kws={"color":"b","edgecolor":"m","linewidth":5,"linestyle":"--","alpha":0.9},
             kde=False,
             fit=norm,
             fit_kws={"color":"y","linewidth":5,"linestyle":"--","alpha":0.9},
            rug=True,
            rug_kws={"color":"k","edgecolor":"m","linewidth":5,"linestyle":"--","alpha":0.9},label="TB")
plt.title("histogram of Restorant total bill",fontsize=20)
plt.xlabel("total bill",fontsize=15)
plt.legend()
plt.show()
```
